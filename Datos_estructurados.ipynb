```python
# prompt: Cual es la aerolinea que \tiene el precio mas bajo

# Primero, creamos una nueva columna con los precios convertidos a float
emily_vuelos['Precio'] = emily_vuelos['Precio'].str.replace('$', '').str.replace(',', '').astype(float)

# Luego, buscamos la aerolínea con el precio más bajo
aerolinea_mas_barata = emily_vuelos[emily_vuelos['Precio'] == emily_vuelos['Precio'].min()]['Aerolinea'].values[0]

# Finalmente, imprimimos la aerolínea con el precio más bajo
print(aerolinea_mas_barata)
```
