```python
#cargar tabla con la info original

#Librerias
import pandas as pd
import sqlite3

datos_productos = pd.read_excel("/content/Copia de Control de Stock mejorado.xlsx", sheet_name="Productos")
datos_ventas = pd.read_excel("/content/Copia de Control de Stock mejorado.xlsx", sheet_name="Ventas")
datos_compras = pd.read_excel("/content/Copia de Control de Stock mejorado.xlsx", sheet_name="Compras")

#visualizamos los primeros registros
datos_productos.head()
```

```python
#creamos la base de datos de interes
#creamos el conector a la base de datos

conn = sqlite3.connect("maspormenos.db")
```

```python
#creamos las tablas de acuerdo a nuestro modelo entidad-relación
#creamos la tabla PRODUCTOS
#definimos la consulta
crear_tabla_producto = """CREATE TABLE IF NOT EXISTS PRODUCTOS(
  CODIGO TEXT PRIMARY KEY,
  PRODUCTO TEXT,
  PRECIOCOMPRA REAL,
  PRECIOVENTA REAL,
  ENTRADAS INTEGER,
  SALIDAS INTEGER,
  STOCKACTUAL INTEGER,
  TALLAXS INTEGER,
  TALLAS INTEGER,
  TALLAM INTEGER,
  TALLAL INTEGER,
  TALLAXL INTEGER,
  TALLA2XL INTEGER,
  TALLA3XL INTEGER

)"""

#creamos un apuntador a la base de datos (cursor)
apuntador = conn.cursor()

#ejecuto el query o la consulta
apuntador.execute(crear_tabla_producto)

#hacemos commit sobre la conexión
conn.commit()
```

```python
#pasamos la información del DataFrame a la tabla PRODUCTOS

#definimos la información que queremos pasar
datos_productos2 = datos_productos[["CODIGO", "PRODUCTO", "PRECIOCOMPRA", "PRECIOVENTA", "ENTRADAS", "SALIDAS", "STOCKACTUAL", "TALLAXS", "TALLAS", "TALLAM", "TALLAL","TALLAXL","TALLA2XL","TALLA3XL"]]

#hacemos el transpaso de la información
datos_productos2.to_sql("PRODUCTOS", conn, if_exists = "append", index = False)
```

```python
#visualizamos la info de la tabla CLIENTE, para efectos de comprobamos
#Se crea la consulta o el query para traer la info de la tabla
consulta = """SELECT * FROM PRODUCTOS"""

#ejecutamos la consulta
apuntador.execute(consulta)

#obtenemos resultados de la consulta
resultado = apuntador.fetchall()

#mostramoss el resultado
#print(resultado)

#mostramos el resultado, que es una lista de tuplas, en forma organizada
for filas in resultado:
  print(filas)
```

```python
#Repetimos los pasos anteriores para las tablas VENTAS y COMPRAS
#para la tabla VENTAS:
crear_tabla_ventas = """CREATE TABLE IF NOT EXISTS VENTAS(
  FECHA DATE,
  CODIGO TEXT,
  PRODUCTO TEXT,
  CANTIDAD INTEGER,
  PRECIOVENTA REAL,
  TOTALVENTA REAL,
  GANANCIA REAL,
  CANAL TEXT,
  LOCALIZACIÓN TEXT,
  TALLA TEXT,
  METODODEPAGO TEXT,
  COSTOENVIO REAL,
  FOREIGN KEY(CODIGO) REFERENCES PRODUCTO("CODIGO"))"""

#creamos un apuntador a la base de datos (cursor)
apuntador = conn.cursor()

#ejecuto el query o la consulta
apuntador.execute(crear_tabla_ventas)

#hacemos commit sobre la conexión
conn.commit()
#definimos la información que queremos pasar
datos_ventas2 = datos_ventas[["FECHA", "CODIGO", "PRODUCTO", "CANTIDAD", "PRECIOVENTA", "TOTALVENTA", "GANANCIA", "CANAL", "LOCALIZACIÓN", "TALLA", "METODODEPAGO", "COSTOENVIO"]]

#hacemos el transpaso de la información
datos_ventas2.to_sql("VENTAS", conn, if_exists = "append", index = False)
```

```python
#Se crea la consulta o el query para traer la info de la tabla
consulta = """SELECT * FROM VENTAS"""

#ejecutamos la consulta
apuntador.execute(consulta)

#obtenemos resultados de la consulta
resultado = apuntador.fetchall()

#mostramoss el resultado
#print(resultado)

#mostramos el resultado, que es una lista de tuplas, en forma organizada
for filas in resultado:
  print(filas)
```

```python
#para la tabla COMPRAS:
crear_tabla_compras = """CREATE TABLE IF NOT EXISTS COMPRAS(
  FECHA DATE,
  CODIGO TEXT,
  PRODUCTO TEXT,
  CANTIDAD INTEGER,
  PRECIOCOMPRA REAL,
  TOTALCOMPRA REAL,
  FOREIGN KEY(CODIGO) REFERENCES PRODUCTO("CODIGO"))"""

#creamos un apuntador a la base de datos (cursor)
apuntador = conn.cursor()

#ejecuto el query o la consulta
apuntador.execute(crear_tabla_compras)

#hacemos commit sobre la conexión
conn.commit()
#definimos la información que queremos pasar
datos_compras2 = datos_compras[["FECHA", "CODIGO", "PRODUCTO", "CANTIDAD", "PRECIOCOMPRA", "TOTALCOMPRA"]]

#hacemos el transpaso de la información
datos_compras2.to_sql("COMPRAS", conn, if_exists = "append", index = False)
```

```python
#Se crea la consulta o el query para traer la info de la tabla
consulta = """SELECT * FROM COMPRAS"""

#ejecutamos la consulta
apuntador.execute(consulta)

#obtenemos resultados de la consulta
resultado = apuntador.fetchall()

#mostramoss el resultado
#print(resultado)

#mostramos el resultado, que es una lista de tuplas, en forma organizada
for filas in resultado:
  print(filas)
```

```python
#guardamos cambios
conn.commit()
conn.close()
```
